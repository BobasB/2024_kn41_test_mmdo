# Розпочинаємо основи тестування
- тести створюються для перевірки правильності роботи основної програми;
- в своїй основі, тести це також програми, або частинка коду яка запускає на виконання основний код та перевіряє правильність його роботи;
- в основному ми будемо зосереджуватись на Юніт тестуванні (пееревірки роботи якось юніта або частини коду) або на Функціональному тестуванні (перевірки чи код працює так як задумувалось);
- всі тести будуть в окремому файлі (можна також розбивати тести на різні файли, і поміщати у окрему папку, але ми так робити не будемо);
- ми не зможемо напряму працювати з класом Меча, написаним у роботі 2, тому ми просто будемо копіювати код, модифікувати його якщо потрібно, та формувати у реальний проект, з використання тестів;
