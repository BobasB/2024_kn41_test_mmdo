## А що таке класи та обєкти?
> Вся ідея ООП полягає в роботі з абстракціями які описують якесь явище або поняття що існує в реальності (для прикладу цифри це є абстракція яка дозволяє нам взнати кільсть чогось, або слова `машина` це абстракція якою позначають кабіну з чотирьома колесами яка може їздити);
> За допомогою програмного коду класів та обєктів ми може описати для компютера що таке певна абстракція (для прикладу, розробника Пайтон вже описали абстрацію чисел, і тому ми можемо користуватись `int` або `float`);
> В класичному вигляді для пояснення класі і обєкту говорять що: `Клас` це є шаблон, блупрінт (схема/креслення) який описує якесь фізичне явище, чи предмет, чи абстракцію яку ми хочем описати для компютера (для прикладу, пере ти як створити машину, малюють її концепт, створюють дизайн, інженери роблять креслення для того щоб описати деталі та як їх складати). По інструкції, яку надає `Клас` ми можемо створити `Обєкт` який є вираженням фізичної сутності того що описує клас. Слід памятати, що обєкти є унікальними і на прикладі машин, є Машина - Меседес С300, і хоча це модель є однаковою однак конкретні машини будуть відрізнятись (кожен обєкт/екземпляр буде мати свій колір, свої параметри двигуна, свій тюнін, зношеність);
